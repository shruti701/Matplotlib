### 1)Install matplotlib and import pyplot.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### 2)Plot a basic line chart.

```python
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
    
```

### 3)Add title and axis labels.

```python
plt.title("My Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

### 4)Change line color, style, and markers.

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,48,46]
```

```python
plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.show()
```

### 5)Plot multiple lines.

```python
plt.plot(x, y)
plt.plot(x, y)
plt.show()
```

### 6)Add a legend.

```python
plt.legend(["Line 1", "Line 2"])
plt.show()
```

### 7)Save the plot.

```python
plt.savefig("plot.png")
plt.show()
```

### 8)Bar chart.

```python
plt.bar(['A', 'B', 'C'], [10, 20, 15])
plt.show()
```

### 9)Histogram.

```python
data = [10, 20, 20, 30, 30, 30, 40, 50, 60, 60]  # Example data
plt.hist(data, bins=5)
plt.show()
```

```python
plt.hist(data,bins=5)
plt.show()
```

### 10)Pie chart

```python
plt.pie([30, 40, 30], labels=['A', 'B', 'C'])
plt.show()
```

### 11)Subplots.

```python
fig, axs = plt.subplots(1, 2)
axs[0].plot(x, y)
axs[1].plot(x, y)
plt.show()
```

### 12)Customize ticks and grid.

```python
plt.xticks(range(0, 10, 2))
plt.grid(True)
plt.show()
```

### 13)Figure size.

```python
plt.figure(figsize=(8, 4))
plt.show()
```

### 14)Scatter plot.

```python
plt.scatter(x, y)
plt.show()
```

### 15)Add annotations.

```python
plt.annotate("Point", (x,y))
plt.show()
```

### 16)Stacked bar chart.

```python
plt.bar(x, y)
plt.bar(x, y, bottom=y)
plt.show()
```

### 17)Change background and axis colors.

```python
ax = plt.gca()
ax.set_facecolor('lightgray')
plt.show()
```

### 18)Twin axes.

```python
ax1 = plt.gca()
ax2 = ax1.twinx()
plt.show()
```

### 19)Use different styles.
