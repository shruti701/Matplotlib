```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,45,56,23,40,21,23]
min_t=[34,35,36,37,38,23,12]
avg_t=[41,43,42,56,76,67,43]
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.show()
```

```python
# set axes lables and chart title
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.show()
```

### Legend

```python
# show legend
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend()
plt.show()
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend(loc='upper right',shadow=True,fontsize='large')
plt.show()
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend(loc='lower left',shadow=True,fontsize='small')
plt.show()
```
