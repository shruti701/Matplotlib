```python
### matplotib Tutorial : pyplot.plot fromat strings and other options
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,48,46]
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'b+--')
plt.show()
```

```python
# same effect as 'r+'format string
plt.plot(x,y,color='red',marker='+',linestyle='',markersize=10)
plt.show()
```

## Markersize

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=10)
plt.show()
```

```python
### alpha property to control transparncy of line chart
```

```python
plt.plot(x,y,'r>',alpha=0.5)
plt.show()
# alpha can be specified 
```
